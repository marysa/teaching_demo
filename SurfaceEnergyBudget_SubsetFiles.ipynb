```python
# ds_1pctco2
amon_vars = ['rlds','rlus','rsds','rsus','rsdscs','rsuscs','ts','tas','rsut','rlut','rsdt','hfls','hfss','ua','va']#,'co2']
lmon_vars = ['lai']
```

```python
path_dict = {}
path_dict['1pctCO2'] = '/glade/collections/cdg/data/CMIP6/CMIP/NCAR/CESM2/1pctCO2/r1i1p1f1/'
# path_dict['ssp585'] = '/glade/collections/cdg/data/CMIP6/ScenarioMIP/NCAR/CESM2/ssp585/r10i1p1f1/'


run_dict  = {}
run_dict['1pctCO2'] = 'v20190425'
# run_dict['ssp585'] = 'v20200528'

```

```python
ds_dict = {}

for run in run_dict.keys():
    print(run)
    ds_dict[run] = {}
    
    ds_list = []
    
    for var in amon_vars:
        ds_temp = xr.open_mfdataset(path_dict[run]+'/Amon/'+var+'/gn/'+run_dict[run]+'/*.nc')
        ds_list.append(ds_temp)
        
    ds_all = xr.merge(ds_list)
    ds_all['rsns'] = ds_all['rsds'] - ds_all['rsus']
#     da_rsns = ds_all['rsds'] - ds_all['rsus']
#     ds_rsns = da_rsns.to_dataset(name='rsns')
#     ds_all = xr.merge([ds_all,ds_rsns])
    
    ds_dict[run]['Amon'] = ds_all
    
    
    ds_list = []
    
    for var in lmon_vars:
        ds_temp = xr.open_mfdataset(path_dict[run]+'/Lmon/'+var+'/gn/'+run_dict[run]+'/*.nc')
        ds_list.append(ds_temp)
        
    ds_dict[run]['Lmon'] = xr.merge(ds_list)
```
